<h1> Pitch Correction prototype based on the Auto-Tune patent </h1>

It is necessary to install all of these dependencies to be able to run the prototype succesfully

```python
import numpy as np
from scipy.io import wavfile

import matplotlib.pyplot as plt
import sounddevice as sd

# PitchCorrection Class
from PitchCorrection import PitchCorrector

from scipy.signal import butter, lfilter, freqz
```

<h2> Mantained synthesized piano note</h2>

First we read the wavfile. All of the wavfiles used are available in the samples folder. First we will test the algorithm with a part of a song that isn't very complex and doesn't have big jumps in notes.

```python
fs, wav = wavfile.read('samples/pianoA.wav')
```

<h3> Original waveform </h3>

Here, the original waveform is plotted as a mean of comparison when the Pitch Corrector has built the tuned output.

```python
plt.plot(wav)
```

```python
plt.plot(wav[0:200000]) # Representation of the first segment in the waveform
```

```python
plt.plot(wav[40000:50000]) # Representation of the highest peak in the first segment
```

```python
plt.plot(wav[48000:50000])
```

<h3> Let's hear the original </h3>

Before we start the pitch correction let's hear the original audio.

```python
sd.play(wav)
```

<h3> Pitch Correction </h3>

Now, we can use our class PitchCorrector to correct the pitch of the original wavfile. 

```python
corrected_wav = PitchCorrector(wav, sensitivity=.05, min_freq=80, max_freq=220)
```

<h3> Visualizing the results </h3>

Now we will plot the resulting audio to compare it to the original and then we will also plot the frequencies found and estimated as a way to see how the pitch detection algorithm performed in a more visual way.

```python
plt.plot(corrected_wav.output)
```

```python
plt.plot(corrected_wav.output[0:200000])
```

```python
plt.plot(corrected_wav.output[40000:50000]) # Representation of the highest peak in the first segment
```

```python
plt.plot(corrected_wav.output[48000:50000]) # Representation of the highest peak in the first segment
```

```python
plt.plot(corrected_wav.real_frequencies)
```

```python
plt.plot(corrected_wav.desired_frequencies)
```

<h2> Let's hear the results </h2>

```python
corrected_wav.play()
```
