<h1> Pitch Correction prototype based on the Auto-Tune patent </h1>

It is necessary to install all of these dependencies to be able to run the prototype succesfully

```python
import numpy as np
from scipy.io import wavfile

import matplotlib.pyplot as plt
import sounddevice as sd

# PitchCorrection Class
from PitchCorrection import PitchCorrector

from scipy.signal import butter, lfilter, freqz
```

For all the tests the same steps will be followed:
   - Read the wavefile
   - Plot the original waveform
   - Listen to the original waveform
   - Use PitchCorrector to build the corrected output
   - Plot the corrected waveform as well as the found frequencies and the corresponding desired frequencies generated
   - Listen to the corrected waveform
   
A more detailed explanation on all these steps is given in the first round of tests.

<h2> First round of tests - Mantained note (A) </h2>

First we read the wavfile. All of the wavfiles used are available in the samples folder. In this first round of tests we will be using the recording of a singer singing a sustained A.

```python
fs, wav = wavfile.read('samples/matained_a.wav')
```

<h3> Original waveform </h3>

If the prototype is succesfull, the resulting waveform shouldn't be much different than the original. Here, the original waveform is plotted as a mean of comparison when the Pitch Corrector has built the tuned output.

```python
plt.plot(wav)
```

```python
plt.plot(wav[0:200000]) # Representation of the first segment in the waveform
```

```python
plt.plot(wav[40000:50000]) # Representation of the highest peak in the first segment
```

```python
plt.plot(wav[48000:50000])
```

<h3> Let's hear the original </h3>

Seeing the detected frequencies and the plotted waveforms is a good way of knowing if the pitch detection and pitch shifting have done their job correctly. However, there is something quite satisfying about hearing the results for yourself. First, we should hear the original (and terrible) rendition.

```python
sd.play(wav)
```

<h3> Pitch Correction </h3>

Now, we can use our class PitchCorrector to correct the pitch of the original wavfile. For this particular song, better results are achieved using a minimum frequency of 80 and a maximum frequency of 220. Since the resulting prototype is quite slow, reducing the range of possible frequencies gives us better results quicker. Also, a sensitivity of .05 is chosen so the system is not too strict.

Further explanations on how the code works are given inside PitchCorrection.py

```python
corrected_wav = PitchCorrector(wav, sensitivity=.05, min_freq=80, max_freq=220)
```

<h3> Visualizing the results </h3>

Now we can see that the waveform hasn't been altered in a significant way. Also, if we plot the found frequencies and the desired frequencies to which the audio has been corrected we can check that the pitch detection and the system that chooses the desired frequency are being accurate. Unfortunatly, not all the found frequencies are correct since further optimizations need to be made in order to increase the accuracy of the system.

```python
plt.plot(corrected_wav.output)
```

```python
plt.plot(corrected_wav.output[0:200000])
```

```python
plt.plot(corrected_wav.output[40000:50000]) # Representation of the highest peak in the first segment
```

```python
plt.plot(corrected_wav.output[48000:50000]) # Representation of the highest peak in the first segment
```

```python
plt.plot(corrected_wav.real_frequencies)
```

```python
plt.plot(corrected_wav.desired_frequencies)
```

```python
corrected_wav.play()
```
